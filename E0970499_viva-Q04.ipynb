```python
#e is short for ethanol
#this code finds out the types of delimiters of ethanol.txt files 
e_d_list = []

for file in glob.glob('*'):
    with open (f'{file}') as f:
        contents = f.readline()
        e_d_list.append(contents)
    
np_e_d_list = np.array(e_d_list)
np.unique(np_e_d_list)

#so based on the output, there is only '\t', ' ' and ','
```

```python
#creating a list of txt files of ethanol, but delimiters and new line spacings have not been deleted
e_list = []

for file in glob.glob('*'):
    with open (f'{file}') as f:
        contents = f.readlines()
        e_list.append(contents)
    
e_1d_list = sum(e_list, [])     #oct_list is a 2d list and needs to be converted to a 1d list to be iterated

#removing \n
e_list_v2 = []
for value in e_1d_list:
    e_list_v2.append(value.replace('\n',''))
    
#deleting delimiters, and splitting the values
e_delimiters = ['\t', ' ', ',']
final_e_list = []
for value in e_list_v2:
    for delimiter in e_delimiters:
        if '#' in value:
            continue
        elif delimiter in value:
            str_value = value.split(delimiter)
            final_value = ([float(i) for i in str_value])
            if final_value not in final_e_list:
                final_e_list.append(final_value)
            else:
                continue
        else:                 #if delimiter is not in value
            continue
```

```python
values_for_e = sum(sorted(final_e_list), [])
values_for_e
```

## Plotting all values

```python
#--------- Generate x and y values --------#
t_x = values_for_t[0::2]
t_y = values_for_t[1::2]
oct_x = values_for_oct[0::2]
oct_y = values_for_oct[1::2]
e_x = values_for_e[0::2]
e_y = values_for_e[1::2]

#------- Plot the data -------#
fig, axes = plt.subplots(nrows=3, ncols=1,
                         figsize=(12, 8), sharey='row')

#------- Subplot 1 -------#
axes[0].set_title('IR Spectra of 1-OCTENE, ETHANOL & TOLUENE')
axes[0].plot(oct_x, oct_y, color='slateblue', label='1-OCTENE')
axes[0].grid(alpha=0.25)
axes[0].fill_between(oct_x, oct_y, 1, color='slateblue', alpha=0.25)
axes[0].legend(loc='lower left', frameon=True)

#------- Subplot 2 -------#
axes[1].set_ylabel('Transmittence')
axes[1].plot(e_x, e_y, color='tab:green', label='ETHANOL')
axes[1].grid(alpha=0.25)
axes[1].fill_between(e_x, e_y, 1, color='tab:green', alpha=0.25)
axes[1].legend(loc='lower left', frameon=True)

#------- Subplot 3 -------#
axes[2].plot(t_x, t_y, color='tab:orange', label='TOLUENE')
axes[2].grid(alpha=0.25)
axes[2].fill_between(t_x, t_y, 1, color='tab:orange', alpha=0.25)
axes[2].legend(loc='lower left', frameon=True)
axes[2].set_xlabel('Wavenumber($cm^{-1}$)')
```
