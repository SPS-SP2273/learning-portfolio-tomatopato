```python
#deleting /n, delimiters, splitting the values, and converting values to float
delimiters = ['\t', ' ', ',']      #define specific delimiters for identification of values

def get_all_values(all_files):
    list_v1 = []
    for file in all_files:                       #creating a list of all txt files 
        with open (f'{file}') as f:
            contents = f.readlines()
            list_v1.append(contents)
    list_v1_1d = sum(list_v1, [])                #list_v1 is a 2d list, must be converted to a 1d list to be iterated
    list_v2 = []
    final_list = []
    for value in list_v1_1d:
        list_v2.append(value.replace('\n',''))
    for value in list_v2:
        for delimiter in delimiters:
            if '#' in value:
                continue
            elif delimiter in value:
                str_value = value.split(delimiter)
                final_value = ([float(i) for i in str_value])
                if final_value not in final_list:
                    final_list.append(final_value)
                else:
                    continue
            else:                 #if delimiter is not in value
                continue
    return sum(sorted(final_list), [])     #final_list is a 2d list, must be converted to a 1d list to be iterated   

values_for_t = get_all_values(glob.glob("*"))
values_for_t
```
